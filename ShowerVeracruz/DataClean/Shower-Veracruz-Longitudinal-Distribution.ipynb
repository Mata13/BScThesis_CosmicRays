**Simulacion de una cascada de Rayos Cosmicos en CORSIKA**

<font color = 'limegreen'>
<font size = 5>
En este notebook se analizan los datos obtenidos de la simulacion de una cascada de particulas usando el programa de CORSIKA, donde el punto de observacion es el Puerto de Veracruz.

<font color = 'coral'>
<font size = 3.5>
CORSIKA es un programa para la simulacion de chubascos de particulas en la atmosfera producidos por Rayos Cosmicos. El programa se puede obtener de manera gratuita, para mas detalles y documentacion consulta la web oficial de <a href=https://www.iap.kit.edu/corsika/79.php>CORSIKA</a>.

<font color = 'slateblue'>
<font size = 3>
Para realizar la simulacion se usaron los siguientes datos:

- Particulas primarias (clave CORSIKA):

    - Proton: 14
    - Nucleo de helio: 402
    - Nucleo de Fe: 5626

- Energia de la particula primaria:
    - $1.0 \times 10^{5}$ eV

- Coordenadas del punto de observacion:

    - 19 10 32 N
    - 96 07 13 W

- Altitud del punto de observacion (nivel del mar) = 0

- Valor del campo magnetico **B**:

    - 39,721.5 nT = 39.7215 &mu;T

- Componentes del campo magnetico **B**:

    - Horizontal ==> **Bx** = 27,030.6 nT = 27.0306 &mu;T
    - Vertical ==> **Bz** = 29,078.3 nT = 29.0783 &mu;

------------------------------------------------------
- Nomenclatura de los archivos out de la simulacion:

    - DAT000001 --> Shower-Veracruz-Proton
    - DAT000002 --> Shower-Veracruz-Helium
    - DAT000003 --> Shower-Veracruz-Fe

    - DAT000004 --> Shower-SierraNegra-Proton
    - DAT000005 --> Shower-SierraNegra-Helium
    - DAT000006 --> Shower-SierraNegra-Fe
------------------------------------------------------

```python
# Importamos las librerias que ocupamos
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
```

Distribucion de particulas en una cascada producida por un proton (RC primario)

```python
# Importamos data set de una simulacion de una cascada producida por un proton
df = pd.read_csv('Shower-proton/Veracruz-Longitudinal-Distribution-Proton.txt', delimiter = ',')
```

```python
# Imprimimos las primeras filas del data set para verificar si cargaron exitosamente
df.head()
```

```python
# Imprimimos las ultimas filas del dataframe para inspeccionar los datos
df.tail()
```

```python
# Imprimimos las dimensiones del data set (rows, columns)
df.shape
```

```python
# Graficamos la distribucion de los rayos gammma producidos en una cascada por un proton 
plt.plot(df['GAMMAS'], df['DEPTH'], label = 'Rayos Gamma', color = 'limegreen')

# Ajustes del grafico
plt.xlabel('Particulas')
plt.ylabel('Profundidad atmosferica [g/$cm^2$]')
plt.title('Distribucion de los Rayos Gamma producidos en una cascada por un proton')
plt.legend()
plt.grid(True)
```

```python
# Graficamos la distribucion de los nucleones producidos en una cascada por un proton
plt.plot(df['NUCLEI'], df['DEPTH'], label = 'nucleones', color = 'indianred')

# Ajustes del grafico
plt.xlabel('Numero de particulas')
plt.ylabel('Profundidad atmosferica [g/$cm^2$]')
plt.title('Distribucion de los nucleones producidos en una cascada por un proton')
plt.legend()
plt.grid(True)
```
