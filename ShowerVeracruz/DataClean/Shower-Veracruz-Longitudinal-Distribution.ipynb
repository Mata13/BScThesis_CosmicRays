**Simulación de una cascada de Rayos Cósmicos en CORSIKA**

<font color = 'limegreen'>
<font size = 5>
En este notebook se analizan los datos obtenidos de la simulación de una cascada de partículas usando el programa de CORSIKA, donde el punto de medición es el Puerto de Veracruz.

<font color = 'coral'>
<font size = 3.5>
CORSIKA es un programa para la simulación de chubascos de partículas en la atmósfera producidos por Rayos Cósmicos. El programa se puede obtener de manera gratuita, para más detalles y documentación consulta la web oficial de <a href=https://www.iap.kit.edu/corsika/79.php>CORSIKA</a>.

<font color = 'slateblue'>
<font size = 3>
Para realizar la simulación se usaron los siguientes datos:

- Partículas primarias (clave CORSIKA):

    - Protón: 14
    - Núcleo de helio: 402
    - Núcleo de Fe: 5626

- Energía de la partícula primaria:
    - $1.0 \times 10^{5}$ eV

- Coordenadas del punto de observación:

    - 19 10 32 N
    - 96 07 13 W

- Altitud del punto de observación (nivel del mar) = 0

- Valor del campo magnético **B**:

    - 39,721.5 nT = 39.7215 &mu;T

- Componentes del campo magnético **B**:

    - Horizontal ==> **Bx** = 27,030.6 nT = 27.0306 &mu;T
    - Vertical ==> **Bz** = 29,078.3 nT = 29.0783 &mu;

------------------------------------------------------
- Nomenclatura de los archivos out de la simulación:

    - DAT000001 --> Shower-Veracruz-Proton
    - DAT000002 --> Shower-Veracruz-Helium
    - DAT000003 --> Shower-Veracruz-Fe

    - DAT000004 --> Shower-SierraNegra-Proton
    - DAT000005 --> Shower-SierraNegra-Helium
    - DAT000006 --> Shower-SierraNegra-Fe

Distribución de partículas en una cascada producida por un protón (RC primario)

```python
# Importamos data set de una simulación de una cascada producida por un protón
df = pd.read_csv('Shower-proton/Veracruz-Longitudinal-Distribution-Proton.txt', delimiter = ',')
```

```python
# Imprimimos las primeras filas del data set para verificar si cargaron exitosamente
df.head()
```

```python
# Imprimimos las ultimas filas del dataframe para inspeccionar los datos
df.tail()
```

```python
# Imprimimos las dimensiones del data set (rows, columns)
df.shape
```

```python
# Graficamos todas las partículas en un mismo lienzo
plt.figure(figsize = (10, 6)) # creamos un lienzo

# plt.plot(df['Gammas'], df['Profundidad'], label = 'Rayos Gamma', color = 'cyan')
plt.plot(df['GAMMAS'], df['DEPTH'], label = '$\gamma$', color = 'tan')
plt.plot(df['POSITRONS'], df['DEPTH'], label = '$e^+$', color = 'chocolate')
plt.plot(df['ELECTRONS'], df['DEPTH'], label = '$e^-$', color = 'turquoise')
plt.plot(df['MU+'], df['DEPTH'], label = '$\mu^+$', color = 'dodgerblue')
plt.plot(df['MU-'], df['DEPTH'], label = '$\mu^-$', color = 'darkorange')
plt.plot(df['HADRONS'], df['DEPTH'], label = 'hadrones', color = 'cyan')
plt.plot(df['CHARGED'], df['DEPTH'], label = 'partículas cargadas', color = 'mediumslateblue')
plt.plot(df['NUCLEI'], df['DEPTH'], label = 'nucleones', color = 'indianred')

plt.xlabel('Partículas secundarias')
plt.ylabel('Profundidad atmosférica [g/$cm^2$]')
plt.title('Distribución de las partículas producidas en una cascada por un protón')
plt.legend()
plt.grid(True)
```
