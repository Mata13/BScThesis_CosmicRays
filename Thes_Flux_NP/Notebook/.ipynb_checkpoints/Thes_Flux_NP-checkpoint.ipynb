**Simulacion del Flujo de Rayos Cosmicos en CORSIKA usando el Cluster LARCAD**

<font color = 'limegreen'>
<font size = 5>
En este notebook se analizan los datos obtenidos de la simulacion del flujo de RC usando el programa de CORSIKA en el cluster de LARCAD,
para ver como cambia el flujo a medida que subimos en altitud desde el nivel del mar hasta 4500 msnm.

<font color = 'slateblue'>
<font size = 3>
Datos utiles para la simulacion:

- Particulas primarias (clave CORSIKA):

    - Proton: 14
    - Nucleo de helio: 402
    - Nucleo de Fe: 5626

Observacion: el programa de CORSIKA se corre en el cluster, los archivos generados en la simulacion se analizan usando Python.
Se van a generar las graficas necesarias para analizar los resultados de esta tesis.

<font color = 'deepskyblue'>
<font size = 5>
Librer√≠as

```python
# Importamos las ibrerias que ocupamos
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import os
```

<font color = 'orchid'>
<font size = 5>
Energia vs particulas producidas.
1) Altutud: 7 msnm

```python
# Obtenemos las ruta del notebook
current_directory = os.getcwd()
print(current_directory)
```

```python
# Importamos el data set para altitud de 7 msnm.
df_Flux_7m = pd.read_csv('../DataClean/CR_Flux_7m.txt', delimiter = ',')
```

```python
# Imprimimos las primeras filas del data set para verificar si cargaron exitosamente
df_Flux_7m.head()
```

```python
# Imprimimos las ultimas filas del dataframe para inspeccionar los datos
df_Flux_7m.tail()
```

```python
# Imprimimos las dimensiones del data set (rows, columns)
df_Flux_7m.shape
```

```python
# Graficamos los positrones y su energia a 7 msnm 
plt.loglog(df_Flux_7m['N_e+'], df_Flux_7m['Energy (GeV)'], label = 'positrones', color = 'limegreen')

# Ajustes del grafico
plt.xlabel('Particulas')
plt.ylabel('Energia [GeV]')
plt.title('Distribucion de la energia de los positrones a 7 msnm')
plt.legend()
plt.grid(True)
```

```python
# Graficamos todas las particulas en un mismo lienzo usando escala logo-log
plt.figure(figsize = (10, 6)) # creamos un lienzo

# Distribucion de la energia de las particulas secundarias a 7 msnm
plt.loglog(df_Flux_7m['Photons'], df_Flux_7m['Energy (GeV)'], label = 'fotones', color = 'limegreen')
plt.loglog(df_Flux_7m['N_e+'], df_Flux_7m['Energy (GeV)'], label = 'positrones', color = 'chartreuse')
plt.loglog(df_Flux_7m['N_e-'], df_Flux_7m['Energy (GeV)'], label = 'electrones', color = 'salmon')
plt.loglog(df_Flux_7m['N_mu+'], df_Flux_7m['Energy (GeV)'], label = 'muones+', color = 'navy')
plt.loglog(df_Flux_7m['N_mu-'], df_Flux_7m['Energy (GeV)'], label = 'muones-', color = 'orchid')
plt.loglog(df_Flux_7m['N_pi0'], df_Flux_7m['Energy (GeV)'], label = 'piones 0', color = 'cyan')
plt.loglog(df_Flux_7m['N_pi+'], df_Flux_7m['Energy (GeV)'], label = 'piones +', color = 'mediumslateblue')
plt.loglog(df_Flux_7m['N_pi-'], df_Flux_7m['Energy (GeV)'], label = 'piones -', color = 'indianred')
plt.loglog(df_Flux_7m['N_n'], df_Flux_7m['Energy (GeV)'], label = 'neutrones', color = 'magenta')
plt.loglog(df_Flux_7m['N_p'], df_Flux_7m['Energy (GeV)'], label = 'protones', color = 'red')
plt.loglog(df_Flux_7m['N_pbar'], df_Flux_7m['Energy (GeV)'], label = 'N_pbar', color = 'chocolate')
plt.loglog(df_Flux_7m['N_others'], df_Flux_7m['Energy (GeV)'], label = 'Otros', color = 'tan')

plt.xlabel('Numero de particulas')
plt.ylabel('Energia (GeV)')
plt.title('Distribucion de la energia de las particulas secundarias a 7 msnm')
plt.legend()
plt.grid(True)
```
